```python
import numpy as np
import os
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from scipy import odr
from scipy import signal
from scipy import fftpack

plt.rcParams["figure.figsize"]=10,5
```

```python
f = 10
t = np.arange(0,1,0.001)
a = 2.
phi = 0.8

signal = a * np.sin(2*np.pi*f*t + phi)
```

```python
N = len(t)

X = fftpack.fft(signal)

fe = (N-1.) / (t[-1] - t[0])

fpos = np.linspace(0., fe/2., np.int(N/2))
Xpos = X[0:np.int(N/2)]

print(fpos)
```

```python
# Plot Data : 

f_m = fpos[np.argmax(abs(Xpos))]
phi_m = np.angle(Xpos)[np.argmax(abs(Xpos))]

fit = np.max(signal) * np.cos(2*np.pi*f_m*t + phi_m )

%matplotlib notebook
plt.figure()

plt.xlabel('$t\ (s)$')
plt.ylabel('$U$')

plt.errorbar( t , signal, color='tomato', fmt = 'o',markersize=4,label='donnÃ©es')
plt.plot(t , fit , color='royalblue')

plt.legend()
plt.show()
```
