```python
import numpy as np
import os
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from scipy import odr
from scipy import signal

plt.rcParams["figure.figsize"]=10,5
```

```python
def my_round(X , DX) :

    if np.log10(DX)>0 : 
        Ndecimal = 1 - np.int(np.ceil(np.abs(np.log10(DX))))
    else : 
        Ndecimal = np.int(np.ceil(np.abs(np.log10(DX))))
    
    if Ndecimal == 0 : 
        DX = np.int(np.around( DX , Ndecimal ))
        X = np.int(np.around( X , Ndecimal ))
    else : 
        DX = np.around( DX , Ndecimal )
        X = np.around( X , Ndecimal )
        
    return X , DX



def Affine(P,x) :
    return P[0] * x + P[1]
```

```python
Deltal = 0.597*10**(-9) # (m) différence de longueur d'onde tabulée
l0 = 589.3*10**(-9) # (m) longueur d'onde moyenne du doublet

e = np.arange(1,2,0.000001) # (mm) Variation typique de l'épaisseur prise par Fruchart

i = np.arange(0,6,0.0001) # (°) le i typique est pris en supposant une figure de 10cm de rayon projetée par une lentille de focale (1m) Soit 6°
```

```python
catinit=catinit.replace(',',' ') #Delimiter 

cat = open('rdy2go.txt','w')
cat.write(catinit)
cat.close()
```

```python
# Lecture du fichier Formate 
Cat = np.genfromtxt('rdy2go.txt',dtype=[('xp',np.float) , ('p',np.float)],skip_header=1)
```

```python
# Incertitude sur la mesure de xp
xp_err_lecture = 0.005  # (mm) Vernier gradué à 0.01
xp_err_estimation = 0.01 # (mm) Incertitude due à l'appréciation de l'anti-coincidence.

xp_err = np.sqrt(xp_err_lecture**2 + xp_err_estimation**2)

#on passe en (m) : 

Cat['xp'] = Cat['xp']*10**(-3)
xp_err = xp_err*10**(-3)
```

```python
## Ajustement  = A * f + B 
myModel = odr.Model(Affine)

mydata = odr.RealData(x=Cat['p'],y=Cat['xp'],sy=xp_err)

myodr = odr.ODR(mydata, myModel,beta0=[0.,0.])

myoutput = myodr.run()
```

```python
# Plot resultat : 

Xrange = 1.5*(np.max(Cat['p']) - np.min(Cat['p']))
Xmin = (np.max(Cat['p']) + np.min(Cat['p']))/2. - 0.5*Xrange
Xmax = (np.max(Cat['p']) + np.min(Cat['p']))/2. + 0.5*Xrange
Xplot = np.arange(Xmin,Xmax,Xrange/100.)

%matplotlib notebook
plt.figure()

#plt.xlim(0,Xmax)

plt.xlabel('$p$')
plt.ylabel('$x_p$')

plt.errorbar( Cat['p'] , Cat['xp'], yerr=xp_err , color='tomato', fmt = 'o',markersize=4, label='données')
plt.plot (Xplot, myoutput.beta[0]*Xplot+myoutput.beta[1], 'royalblue', label='$x_p \, = \, A*p \, + \, B $')


plt.legend()
plt.show()
```

```python
#### Déduction de \Delta \lambda :

###############Valeur de lambda0 : #########
l0 = 589.3 * 10**(-9) # lambda0 en (m) 
l0_err = 0.1 * 10**(-9) # l'incertitude sur sa mesure
############################################

Dl = -l0**2/(2.*myoutput.beta[0])

Dl_err = Dl * np.sqrt( (2*l0_err/l0)**2 + (myoutput.sd_beta[0]/myoutput.beta[0])**2 )


Dl_round = my_round(Dl*10**9,Dl_err*10**9) #(nm)

print("la difference de longueur d'onde est de : ")
print(str(Dl_round[0])+' +/- '+str(Dl_round[1])+' nm')
print('')
print('Pour une valeur tabulée de 0.597 nm ')
```

```python
### Aplication au doublet du sodium : 

i = np.arange(0,2,0.00001)
degtorad = np.pi/180.
#Dphi2 = Dphi * (589.2937-0.597/2.)/(589.2937+0.597/2.)

l1 = (589.2937-0.597/2.)*10**(-9)
l2 = (589.2937+0.597/2.)*10**(-9)
e = 0.004 

%matplotlib notebook
plt.figure()

plt.subplot(211)

plt.ylabel('$I/I_\mathrm{max}$')


R=0.9
FP1 = 1/(1 + (4*R/(1-R)**2) * np.sin(2*np.pi*e*np.cos(i*degtorad)/l1)**2)
FP2 = 1/(1 + (4*R/(1-R)**2) * np.sin(2*np.pi*e*np.cos(i*degtorad)/l2)**2)
FP = FP1/np.max(FP1) + FP2/np.max(FP2)

plt.plot (i, FP/np.max(FP) , 'royalblue', label='F-P : R='+str(R))


plt.plot (i, FP1/np.max(FP) , 'aquamarine', linestyle=':', label='$\lambda_1$' )
plt.plot (i, FP2/np.max(FP)  , 'lime', linestyle=':', label='$\lambda_2$')



plt.gca().set_xticklabels([]) # Hide Y ticks
plt.legend(loc='upper left')



plt.subplot(212)

plt.ylabel('$I/I_\mathrm{max}$')
plt.xlabel('$i\degree$')


Mich1 = (1/2.)*(1+np.cos(4*np.pi*e*np.cos(i*degtorad)/l1))
Mich2 = (1/2.)*(1+np.cos(4*np.pi*e*np.cos(i*degtorad)/l2))

Michelson = Mich1 + Mich2


plt.plot (i, Michelson/np.max(Michelson) , 'tomato', label='Michelson')

plt.plot (i, Mich1/np.max(Michelson) , 'darkred', linestyle=':', label='$\lambda_1$' )
plt.plot (i, Mich2/np.max(Michelson)  , 'peru', linestyle=':', label='$\lambda_2$')


plt.ylim(0,1.1)
plt.legend(loc='upper left')
plt.subplots_adjust(wspace=0, hspace=0) # Ajuster les positions des subplots


plt.plot()
```
